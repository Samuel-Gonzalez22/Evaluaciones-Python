```python
# Evaluaci√≥n 6 - Sistema de Reconocimiento Autom√°tico de Placas (ANPR)

# Importaci√≥n y configuraci√≥n de dependencias
import os
import sys
#--------------------------------------------------------------
# Instalaci√≥n de Tesseract (OCR) y sus herramientas
os.system('apt-get install -y tesseract-ocr > /dev/null 2>&1')
os.system('pip install pytesseract > /dev/null 2>&1')
os.system('pip install pytz > /dev/null 2>&1')
#--------------------------------------------------------------

# M√≥dulos esenciales para el proyecto
import cv2
import numpy as np
import pandas as pd
import pytesseract
from datetime import datetime
from google.colab.patches import cv2_imshow
from google.colab import files
from IPython.display import display, Javascript, HTML
from base64 import b64decode
import io
from PIL import Image as PILImage
import re
import time
import pytz

# AJUSTES DEL SISTEMA
NOMBRE_BD_CSV = "registro_placas_vehiculos.csv"
DIRECTORIO_IMAGENES = "imagenes_placas_detectadas"
CONFIG_OCR_BASE = r'--oem 3 --psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'

# Variables de estado
bandera_activa = True
registros_temporales = 0

# GESTI√ìN DE DATOS (CSV)

def configurar_entorno_y_datos():
    """Inicializa la carpeta de im√°genes y el archivo de registro CSV."""
    if not os.path.exists(DIRECTORIO_IMAGENES):
        os.makedirs(DIRECTORIO_IMAGENES)

    if os.path.exists(NOMBRE_BD_CSV):
        # Carga el marco de datos existente
        return pd.read_csv(NOMBRE_BD_CSV)
    else:
        # Crea y guarda un nuevo marco de datos
        columnas = ['Fecha_Captura', 'Hora_Captura', 'Placa_Identificada', 'Nivel_Confianza', 'Nombre_Archivo_Imagen']
        df = pd.DataFrame(columns=columnas)
        df.to_csv(NOMBRE_BD_CSV, index=False)
        return df

def registrar_resultado(placa, confianza, nombre_imagen):
    """A√±ade un nuevo registro de placa al archivo CSV con hora de Colombia."""
    df = pd.read_csv(NOMBRE_BD_CSV)

    zona_horaria = pytz.timezone('America/Bogota')
    momento_local = datetime.now(zona_horaria)

    entrada_nueva = {
        'Fecha_Captura': momento_local.strftime('%Y-%m-%d'),
        'Hora_Captura': momento_local.strftime('%H:%M:%S'),
        'Placa_Identificada': placa,
        'Nivel_Confianza': f"{confianza}%",
        'Nombre_Archivo_Imagen': nombre_imagen
    }

    df = pd.concat([df, pd.DataFrame([entrada_nueva])], ignore_index=True)
    # Persistencia de los datos en disco
    df.to_csv(NOMBRE_BD_CSV, index=False)
    return df

# ALGORITMOS DE VISI√ìN POR COMPUTADORA

def preparacion_de_imagen(imagen):
    """
    Aplica filtros iniciales: reducci√≥n de ruido, detecci√≥n de bordes y
    ajustes morfol√≥gicos para mejorar la detecci√≥n de contornos irregulares.
    """
    alto, ancho = imagen.shape[:2]
    # Redimensionamiento para velocidad
    if ancho > 1280:
        factor_escala = 1280 / ancho
        imagen = cv2.resize(imagen, None, fx=factor_escala, fy=factor_escala, interpolation=cv2.INTER_AREA)

    # Conversi√≥n a escala de grises
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    # Filtro bilateral para reducir ruido manteniendo bordes
    gris_filtrada = cv2.bilateralFilter(gris, 9, 75, 75)
    # Detecci√≥n de bordes Canny
    bordes = cv2.Canny(gris_filtrada, 50, 150)

    # CR√çTICO: Cierre morfol√≥gico para unir el marco dibujado o los bordes dispersos
    # Esto ayuda a que el algoritmo vea un solo contorno rectangular grande.
    kernel_close = np.ones((3, 3), np.uint8)
    bordes = cv2.morphologyEx(bordes, cv2.MORPH_CLOSE, kernel_close)

    return imagen, gris_filtrada, bordes

def identificar_contorno_placa(imagen_bordes):
    """Localiza el contorno con forma y proporci√≥n de una placa de veh√≠culo."""
    contornos, _ = cv2.findContours(imagen_bordes.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    # Ordenar por √°rea y tomar los 10 m√°s grandes
    contornos_principales = sorted(contornos, key=cv2.contourArea, reverse=True)[:10]

    candidatos_validos = []

    for c in contornos_principales:
        perimetro = cv2.arcLength(c, True)
        aproximacion = cv2.approxPolyDP(c, 0.02 * perimetro, True)

        if len(aproximacion) == 4: # Buscar formas de 4 lados (rect√°ngulos/trapecios)
            x, y, w, h = cv2.boundingRect(aproximacion)
            relacion_aspecto = w / float(h)
            area = w * h

            # CR√çTICO: Criterios heur√≠sticos MUY permisivos para placas de papel/manuales
            # Rango de aspecto m√°s amplio (1.5 a 6.5) y √°rea m√≠nima reducida (800)
            if (1.5 <= relacion_aspecto <= 6.5 and w > 40 and h > 10 and area > 800):
                candidatos_validos.append((aproximacion, area))

    if candidatos_validos:
        # Elegir el contorno m√°s grande y con mejor aspecto
        candidatos_validos.sort(key=lambda x: x[1], reverse=True)
        return candidatos_validos[0][0]

    return None

def optimizar_region_para_ocr(recorte_placa):
    """Prepara la imagen de la placa para mejorar la lectura con Tesseract."""
    # Aumentar la escala 3x para mejorar el reconocimiento de caracteres peque√±os
    placa_optimizada = cv2.resize(recorte_placa, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)

    if len(placa_optimizada.shape) == 3:
        placa_optimizada = cv2.cvtColor(placa_optimizada, cv2.COLOR_BGR2GRAY)

    # Ecualizaci√≥n adaptativa de histograma (CLAHE) para mejor contraste local
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    placa_optimizada = clahe.apply(placa_optimizada)
    # Reducci√≥n de ruido avanzada
    placa_optimizada = cv2.fastNlMeansDenoising(placa_optimizada, h=10)

    # CR√çTICO: Aplicar un filtro de afilado (Sharpening) para resaltar los trazos de marcador
    kernel_afilado = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    placa_optimizada = cv2.filter2D(placa_optimizada, -1, kernel_afilado)

    # Umbralizaci√≥n Binaria de Otsu para convertir a blanco y negro
    _, placa_optimizada = cv2.threshold(placa_optimizada, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Operaci√≥n morfol√≥gica para unir posibles fragmentos de caracteres
    kernel = np.ones((2, 2), np.uint8)
    placa_optimizada = cv2.morphologyEx(placa_optimizada, cv2.MORPH_CLOSE, kernel)

    return placa_optimizada

def extraccion_y_votacion_ocr(imagen_original, contorno_placa_candidata):
    """Recorta, procesa y extrae el texto usando m√∫ltiples configuraciones de Tesseract."""
    if contorno_placa_candidata is None:
        return None, None, 0

    # Crear una m√°scara para recortar la regi√≥n exacta
    mascara = np.zeros(imagen_original.shape[:2], np.uint8)
    cv2.drawContours(mascara, [contorno_placa_candidata], 0, 255, -1)

    (y, x) = np.where(mascara == 255)
    (y1, x1) = (np.min(y), np.min(x))
    (y2, x2) = (np.max(y), np.max(x))
    recorte_placa = imagen_original[y1:y2+1, x1:x2+1]

    placa_procesada = optimizar_region_para_ocr(recorte_placa)

    resultados_ocr = []

    # Intento 1: PSM 7 (Tratamiento como l√≠nea simple) - Configuraci√≥n m√°s com√∫n
    texto_p7 = pytesseract.image_to_string(placa_procesada, config=CONFIG_OCR_BASE)
    limpio_p7 = normalizar_y_validar_texto(texto_p7)
    if limpio_p7:
        resultados_ocr.append(limpio_p7)

    # Intento 2: PSM 8 (Tratamiento como palabra simple) - Alternativa √∫til
    config_p8 = r'--oem 3 --psm 8 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    texto_p8 = pytesseract.image_to_string(placa_procesada, config=config_p8)
    limpio_p8 = normalizar_y_validar_texto(texto_p8)
    if limpio_p8:
        resultados_ocr.append(limpio_p8)

    # Intento 3: Inversi√≥n de imagen (√∫til para placas oscuras o sucias)
    placa_invertida = cv2.bitwise_not(placa_procesada)
    texto_inv = pytesseract.image_to_string(placa_invertida, config=CONFIG_OCR_BASE)
    limpio_inv = normalizar_y_validar_texto(texto_inv)
    if limpio_inv:
        resultados_ocr.append(limpio_inv)

    # Sistema de Votaci√≥n: Selecciona el resultado m√°s frecuente para la confianza
    if resultados_ocr:
        texto_final = max(set(resultados_ocr), key=resultados_ocr.count)
        frecuencia = resultados_ocr.count(texto_final)
        confianza_calculada = int((frecuencia / len(resultados_ocr)) * 100)
        return texto_final, placa_procesada, confianza_calculada

    return None, placa_procesada, 0

def normalizar_y_validar_texto(texto):
    """Limpia el texto, aplica correcciones heur√≠sticas y valida el formato de placa."""
    if not texto:
        return None

    # Eliminar caracteres no alfanum√©ricos y pasar a may√∫sculas
    texto = re.sub(r'[^A-Z0-9]', '', texto.upper())

    # Mapa de correcciones comunes (letras que parecen n√∫meros)
    correcciones = {
        'O': '0', # O may√∫scula por cero
        'I': '1', # I may√∫scula por uno
    }

    # Asume que los √∫ltimos 3-4 caracteres son num√©ricos y corrige all√≠
    if len(texto) >= 5:
        parte_letras = texto[:-3]
        parte_numeros = texto[-3:]

        for original, correccion in correcciones.items():
            parte_numeros = parte_numeros.replace(original, correccion)

        texto = parte_letras + parte_numeros

    # Validaci√≥n final del largo de la placa
    if 5 <= len(texto) <= 8:
        return texto

    return None

def analizar_y_extraer_placa(imagen):
    """Funci√≥n principal que orquesta el procesamiento de la imagen."""
    print("\nüîç Iniciando an√°lisis de imagen...")
    tiempo_inicio = time.time()

    imagen_base, imagen_gris, imagen_bordes = preparacion_de_imagen(imagen)
    contorno_placa = identificar_contorno_placa(imagen_bordes)

    if contorno_placa is not None:
        print("‚úì Contorno de placa detectado")

        imagen_resultado = imagen.copy()
        cv2.drawContours(imagen_resultado, [contorno_placa], -1, (0, 255, 0), 3) # Dibujar contorno

        print("Imagen con detecci√≥n:")
        cv2_imshow(imagen_resultado)

        placa_final, region_optimizada, confianza = extraccion_y_votacion_ocr(imagen_base, contorno_placa)

        if region_optimizada is not None:
            print("\nRegi√≥n de la placa despu√©s de optimizaci√≥n:")
            cv2_imshow(region_optimizada)

        if placa_final:
            print(f"\n PLACA IDENTIFICADA: {placa_final}")
            print(f"Confianza: {confianza}%")
            print(f"Tiempo de proceso: {time.time() - tiempo_inicio:.2f} segundos")
            return placa_final, imagen_resultado, confianza
        else:
            print("\n El OCR fall√≥ al leer el texto.")
            return None, imagen_resultado, 0
    else:
        print("\n No se pudo identificar ning√∫n contorno v√°lido de placa.")
        return None, imagen, 0

# INTERACCI√ìN Y CAPTURA

def capturar_desde_webcam():
    """Captura una imagen usando el acceso JS a la c√°mara web de Colab."""
    print("\nAbriendo la c√°mara web del dispositivo...")
    print("Presiona el bot√≥n para tomar la foto.")

    js_script = """
    async function iniciarCaptura() {
      const contenedor = document.createElement('div');
      const video = document.createElement('video');
      const canvas = document.createElement('canvas');
      const boton_capturar = document.createElement('button');

      contenedor.style.textAlign = 'center';
      boton_capturar.textContent = 'CAPTURAR PLACA';
      boton_capturar.style.cssText = 'font-size: 20px; padding: 15px 30px; margin: 10px; background-color: #007bff; color: white; border: none; border-radius: 5px; cursor: pointer;';

      video.style.maxWidth = '100%';
      video.style.border = '3px solid #007bff';

      contenedor.appendChild(video);
      contenedor.appendChild(document.createElement('br'));
      contenedor.appendChild(boton_capturar);
      document.body.appendChild(contenedor);

      // CORRECCI√ìN CR√çTICA: Se a√±ade 'facingMode: environment' para resolver problemas
      // de acceso y permisos de c√°mara en el entorno de Colab/navegador.
      const stream = await navigator.mediaDevices.getUserMedia({
        video: {
            width: {ideal: 1280},
            height: {ideal: 720},
            facingMode: 'environment' // Intenta forzar la c√°mara principal/trasera
        }
      });
      video.srcObject = stream;
      await video.play();

      await new Promise(resuelve => {
        boton_capturar.onclick = () => resuelve();
      });

      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);

      stream.getTracks().forEach(track => track.stop());
      contenedor.remove();

      return canvas.toDataURL('image/jpeg', 0.95);
    }
    """

    try:
        display(Javascript(js_script))
        from google.colab import output
        datos_imagen = output.eval_js('iniciarCaptura()')

        binario = b64decode(datos_imagen.split(',')[1])
        img_pil = PILImage.open(io.BytesIO(binario))
        array_imagen = np.array(img_pil)
        # Convertir RGB (PIL) a BGR (OpenCV)
        array_imagen = cv2.cvtColor(array_imagen, cv2.COLOR_RGB2BGR)

        return array_imagen
    except Exception as e:
        # Aqu√≠ capturamos el error si el usuario no permite la c√°mara
        if "NotFoundError" in str(e) or "NotAllowedError" in str(e):
             print("\n!!! ERROR DE ACCESO A C√ÅMARA: Aseg√∫rate de PERMITIR el uso de la c√°mara en el navegador.")
        else:
            print(f"!!! Error general durante la captura: {e}")
        return None

# INTERFAZ DE USUARIO Y BUCLE PRINCIPAL

def iniciar_sistema_anpr():
    """Bucle principal de ejecuci√≥n del sistema."""
    global registros_temporales

    print("      SISTEMA DE RECONOCIMIENTO AUTOM√ÅTICO DE PLACAS (ANPR)")

    configurar_entorno_y_datos()

    while True:
        print("\n" + "=" * 70)
        print("      MEN√ö PRINCIPAL")
        print("\n   Selecciona una acci√≥n:\n")
        print("    1.  Capturar placa con c√°mara web")
        print("    2.  Cargar imagen desde disco")
        print("    3.  Visualizar registros (CSV)")
        print("    4.  Modificar o eliminar un registro")
        print("    5.  Descargar CSV de registros")
        print("    6.  Salir del sistema")

        try:
            opcion = input("\n> Ingresa tu opci√≥n (1-6): ").strip()

            if opcion == '1':
                print("      OPCI√ìN 1: CAPTURA EN VIVO")

                imagen_capturada = capturar_desde_webcam()

                if imagen_capturada is not None:
                    placa, imagen_proceso, confianza = analizar_y_extraer_placa(imagen_capturada)

                    if placa:
                        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
                        nombre_imagen = f"captura_{timestamp}.jpg"
                        ruta_imagen = os.path.join(DIRECTORIO_IMAGENES, nombre_imagen)
                        cv2.imwrite(ruta_imagen, imagen_proceso)

                        registrar_resultado(placa, confianza, nombre_imagen)
                        registros_temporales += 1

                        print(f"\n[REGISTRO EXITOSO] Entrada #{registros_temporales}")
                        print(f"  Placa: {placa} | Confianza: {confianza}%")
                    else:
                        print("\n[FALLO] No se pudo extraer la placa con suficiente calidad.")

                input("\n[Presiona ENTER para continuar...]")

            elif opcion == '2':
                print("      OPCI√ìN 2: CARGA DE ARCHIVO")
                print("\nSube una imagen de placa desde tu PC...")

                archivos_subidos = files.upload()

                if archivos_subidos:
                    for nombre_archivo in archivos_subidos.keys():
                        print(f"\nProcesando archivo: {nombre_archivo}")

                        datos_binarios = archivos_subidos[nombre_archivo]
                        imagen = cv2.imdecode(np.frombuffer(datos_binarios, np.uint8), cv2.IMREAD_COLOR)

                        placa, imagen_proceso, confianza = analizar_y_extraer_placa(imagen)

                        if placa:
                            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
                            nombre_imagen = f"archivo_{timestamp}.jpg"
                            ruta_imagen = os.path.join(DIRECTORIO_IMAGENES, nombre_imagen)
                            cv2.imwrite(ruta_imagen, imagen_proceso)

                            registrar_resultado(placa, confianza, nombre_imagen)
                            registros_temporales += 1

                            print(f"\n[REGISTRO EXITOSO] Entrada #{registros_temporales}")
                            print(f"  Placa: {placa} | Confianza: {confianza}%")
                        else:
                            print("\n[FALLO] No se pudo extraer la placa.")

                input("\n[Presiona ENTER para continuar...]")

            elif opcion == '3':
                print("      OPCI√ìN 3: VISUALIZAR REGISTROS")

                if os.path.exists(NOMBRE_BD_CSV):
                    df_datos = pd.read_csv(NOMBRE_BD_CSV)
                    if len(df_datos) > 0:
                        display(df_datos)
                        print(f"\nTotal de registros hist√≥ricos: {len(df_datos)}")

                        if 'Nivel_Confianza' in df_datos.columns:
                            df_datos['Confianza_num'] = df_datos['Nivel_Confianza'].str.replace('%', '').astype(float)
                            promedio = df_datos['Confianza_num'].mean()
                            print(f"Confianza media de las detecciones: {promedio:.1f}%")
                    else:
                        print("[INFO] La base de datos CSV est√° vac√≠a.")
                else:
                    print("[ERROR] El archivo de registro CSV no existe.")

                input("\n[Presiona ENTER para continuar...]")

            elif opcion == '4':
                print("      OPCI√ìN 4: EDICI√ìN DE DATOS")

                if os.path.exists(NOMBRE_BD_CSV):
                    df_datos = pd.read_csv(NOMBRE_BD_CSV)
                    if len(df_datos) > 0:
                        print("Mostrando los √∫ltimos 10 registros para referencia:")
                        display(df_datos.tail(10))

                        print("\nSelecciona la operaci√≥n:\n")
                        print("  1. Corregir una Placa (por su valor actual)")
                        print("  2. Eliminar una fila (por su n√∫mero de √≠ndice)")
                        print("  3. Cancelar")

                        sub_opcion = input("\n> Opci√≥n (1-3): ").strip()

                        if sub_opcion == '1':
                            placa_erronea = input("\n> Ingresa la placa INCORRECTA a buscar: ").strip().upper()

                            indices = df_datos[df_datos['Placa_Identificada'] == placa_erronea].index.tolist()

                            if indices:
                                print(f"Se encontraron {len(indices)} coincidencias con '{placa_erronea}'")
                                placa_correcta = input("> Ingresa la placa CORRECTA: ").strip().upper()

                                if placa_correcta and 5 <= len(placa_correcta) <= 8:
                                    if input(f"\n¬øAplicar cambio '{placa_erronea}' -> '{placa_correcta}'? (si/no): ").lower() == 'si':
                                        df_datos.loc[indices, 'Placa_Identificada'] = placa_correcta
                                        df_datos.to_csv(NOMBRE_BD_CSV, index=False)
                                        print(f"\n[√âXITO] {len(indices)} registro(s) corregido(s).")
                                    else:
                                        print("[INFO] Operaci√≥n cancelada.")
                                else:
                                    print("[ERROR] Placa correcta inv√°lida (5-8 caracteres).")
                            else:
                                print(f"[INFO] No se encontr√≥ la placa '{placa_erronea}'.")

                        elif sub_opcion == '2':
                            indice_fila = input("\n> Ingresa el N√öMERO DE √çNDICE de la fila a eliminar: ").strip()

                            try:
                                idx = int(indice_fila)
                                if 0 <= idx < len(df_datos):
                                    if input(f"\n¬øConfirmar eliminaci√≥n de la fila {idx}? (si/no): ").lower() == 'si':
                                        df_datos = df_datos.drop(idx).reset_index(drop=True)
                                        df_datos.to_csv(NOMBRE_BD_CSV, index=False)
                                        print("\n[√âXITO] Registro eliminado.")
                                    else:
                                        print("[INFO] Eliminaci√≥n cancelada.")
                                else:
                                    print("[ERROR] N√∫mero de √≠ndice fuera de rango.")
                            except ValueError:
                                print("[ERROR] Entrada inv√°lida. Debe ser un n√∫mero entero.")
                        else:
                            print("[INFO] Operaci√≥n de edici√≥n cancelada.")
                    else:
                        print("[INFO] No hay registros para modificar.")
                else:
                    print("[ERROR] El archivo de registro CSV no existe.")

                input("\n[Presiona ENTER para continuar...]")

            elif opcion == '5':
                print("      OPCI√ìN 5: DESCARGAR CSV")

                if os.path.exists(NOMBRE_BD_CSV):
                    df_datos = pd.read_csv(NOMBRE_BD_CSV)
                    print(f"Total de registros a descargar: {len(df_datos)}")
                    if len(df_datos) > 0:
                        display(df_datos.tail(5))

                    files.download(NOMBRE_BD_CSV)
                else:
                    print("[ERROR] No hay archivo CSV para descargar.")

                input("\n[Presiona ENTER para continuar...]")

            elif opcion == '6':
                print("      SALIENDO DEL SISTEMA ANPR...")
                print(f"\nSesi√≥n finalizada. Registros procesados: {registros_temporales}")

                if os.path.exists(NOMBRE_BD_CSV):
                    df_datos = pd.read_csv(NOMBRE_BD_CSV)
                    print(f"Total hist√≥rico en base de datos: {len(df_datos)}")

                print("\n" + "=" * 70 + "\n")
                break

            else:
                print("\n[ADVERTENCIA] Opci√≥n no v√°lida. Intenta de nuevo.")
                time.sleep(1)

        except KeyboardInterrupt:
            print("\n\n[INTERRUPCI√ìN] Sistema detenido por el usuario.")
            break
        except Exception as e:
            print(f"\n[ERROR CR√çTICO] Ocurri√≥ un fallo inesperado: {e}")
            input("\n[Presiona ENTER para continuar...]")


print("\n[INFO] M√≥dulos cargados.")
print("Iniciando interfaz principal...")
time.sleep(1)

if __name__ == '__main__':
    iniciar_sistema_anpr()
```
